```python
import re
import pandas as pd
```

```python
file = open("WhatsApp Chat with BIAS 25' Unofficial.txt", "r", encoding='utf-8')
```

## Preprocessing

```python
data = file.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Correct format string
format_str = "%m/%d/%y, %H:%M - "

# Parsing the date string
df['message_date'] = pd.to_datetime(df['message_date'], format=format_str, dayfirst=False)

# Rename column
df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
# separate users and messages
users = []
messages = []
for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)
    if entry[1:]:     #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns = ['user_message'], inplace = True)

df.head()
```

```python
df['date']
```

```python
df['year'] = df['date'].dt.year
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df['day'] = df['date'].dt.day
```

```python
df['hour'] = df['date'].dt.hour
```

```python
df['minute'] = df['date'].dt.minute
```

### Weekly Acitivity Heatmap

```python
df.head()
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + '-' + str('00'))
    elif hour == 0:
        period.append(str('00') + '+' + str(hour+1))
    else:
        period.append(str(hour) + '-' + str(hour+1))
```

```python
df['period'] = period
```

```python
df.sample(5)
```

```python
df.pivot_table(index='day_name', columns='period', values='message', aggfunc='count').fillna(0)
```

```python
import seaborn as sns
plt.figure(figsize=(20, 6))
sns.heatmap(df.pivot_table(index='day_name', columns='period', values='message', aggfunc='count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```
